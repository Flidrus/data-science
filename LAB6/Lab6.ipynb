```python
import tensorflow as tf

# CREATE
scalar_value = tf.constant(20)

# READ
print(f"Initial scalar value: {scalar_value.numpy()}")

# UPDATE (створюємо новий тензор на основі існуючого)
modified_scalar = tf.add(scalar_value, 10)
print(f"Modified scalar value: {modified_scalar.numpy()}")

# DELETE (присвоюємо значення None для видалення посилання)
scalar_value = None
print(f"Scalar after deletion: {scalar_value}")
```

```python
import tensorflow as tf

# CREATE
vector_data = tf.constant([10, 20, 30, 40, 50])

# READ
print(f"Original vector: {vector_data.numpy()}")

# UPDATE (множення кожного елемента на 3)
transformed_vector = vector_data * 3
print(f"Transformed vector: {transformed_vector.numpy()}")

# DELETE (обнулення посилання)
vector_data = None
print(f"Vector after deletion: {vector_data}")
```

```python
import tensorflow as tf

# CREATE
matrix_data = tf.constant([[7, 14], [21, 28], [35, 42]])

# READ
print(f"Initial matrix:\n{matrix_data.numpy()}")

# UPDATE (віднімання 5 від кожного елемента)
modified_matrix = matrix_data - 5
print(f"Modified matrix:\n{modified_matrix.numpy()}")

# DELETE (видаляємо посилання на матрицю)
matrix_data = None
print(f"Matrix after deletion: {matrix_data}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Генерація випадкових даних для регресії
x_data = np.linspace(-2, 2, 120)  # Вхідні дані з більшою кількістю точок
y_data = 4 * x_data - 1 + np.random.normal(0, 0.2, 120)  # Додаємо більший шум

# Конвертація у тензори TensorFlow
x_tensor = tf.constant(x_data, dtype=tf.float32)
y_tensor = tf.constant(y_data, dtype=tf.float32)

# Виведення перших 7 значень тензорів
print("x_tensor (перші 7 значень):", x_tensor[:7].numpy())
print("y_tensor (перші 7 значень):", y_tensor[:7].numpy())

# Побудова графіку
plt.figure(figsize=(8, 5))
plt.plot(x_data, y_data, 'go', label='Generated Data')  # Використовуємо зелений колір

# Додавання деталей до графіку
plt.title("Synthetic Regression Data")
plt.xlabel("Input Feature")
plt.ylabel("Target Value")
plt.legend(loc='best')

# Показ графіку
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Генерація двокласових даних із центрами у різних координатах
data_points = np.random.randn(250, 2)  
labels = np.zeros(250)  
data_points[125:] += [2, -2]  # Зміщення другої групи у протилежному напрямку
labels[125:] = 1  # Встановлення класу для другої групи

# Конвертація у тензори TensorFlow
data_tensor = tf.constant(data_points, dtype=tf.float32)
label_tensor = tf.constant(labels, dtype=tf.float32)

# Виведення перших 6 елементів
print("Data Tensor (перші 6 точок):", data_tensor[:6].numpy())
print("Label Tensor (перші 6 міток):", label_tensor[:6].numpy())

# Побудова графіку для візуалізації
plt.figure(figsize=(7, 7))
plt.scatter(data_points[:125, 0], data_points[:125, 1], color='purple', label='Class A')
plt.scatter(data_points[125:, 0], data_points[125:, 1], color='red', label='Class B')
plt.title("Generated Classification Dataset")
plt.xlabel("Feature X1")
plt.ylabel("Feature X2")
plt.legend()
plt.grid(True)
plt.show()
```
