```python
import numpy as np

# Створення набору даних (Create)

# 1. Створення лінійного масиву з чисел від 1 до 10
linear_array = np.arange(1, 11)
print("Лінійний масив:", linear_array)

# 2. Створення масиву з 10 випадкових чисел
random_array = np.random.random(10)
print("Випадковий масив:", random_array)

# Читання даних (Read)

# 3. Отримання одного елемента (наприклад, третього елемента лінійного масиву)
element = linear_array[2]
print("Третій елемент лінійного масиву:", element)

# 4. Отримання частини масиву (наприклад, елементів з 2 по 4 індекси)
slice_array = linear_array[2:5]
print("Зріз з індексами 2, 3, 4:", slice_array)

# Оновлення даних (Update)

# 5. Оновлення першого елемента лінійного масиву
linear_array[0] = 100
print("Після оновлення першого елемента:", linear_array)

# 6. Оновлення другого і третього елементів
linear_array[1:3] = [200, 300]
print("Після оновлення другого і третього елементів:", linear_array)

# Видалення даних (Delete)

# 7. Видалення третього елемента (з індексом 2)
new_array = np.delete(linear_array, 2)
print("Після видалення третього елемента:", new_array)

# 8. Видалення першого і другого елементів
new_array = np.delete(linear_array, [0, 1])
print("Після видалення першого і другого елементів:", new_array)
```

```python
import pandas as pd
import numpy as np

# 1. Створення DataFrame з масиву NumPy
array_data = np.random.randint(1, 100, size=(5, 3))  # Масив 5x3 з випадковими числами
df_from_array = pd.DataFrame(array_data, columns=['Rating', 'Votes', 'Revenue'])  # DataFrame з іменами колонок
print("DataFrame з масиву NumPy:")
print(df_from_array, "\n")

# 2. Створення DataFrame зі словника
data_dict = {
    'Film': ['Inception', 'The Matrix', 'Interstellar', 'The Godfather', 'Pulp Fiction'],
    'Year': [2010, 1999, 2014, 1972, 1994],
    'Director': ['Christopher Nolan', 'Lana Wachowski', 'Christopher Nolan', 'Francis Ford Coppola', 'Quentin Tarantino']
}
df_from_dict = pd.DataFrame(data_dict)
print("DataFrame зі словника:")
print(df_from_dict, "\n")

# 3. Відображення перших рядків (head)
print("Перші три рядки DataFrame зі словника:")
print(df_from_dict.head(3), "\n")  # Виводимо 3 перші рядки

# 4. Опис статистики (describe)
print("Опис статистики DataFrame з масиву NumPy:")
print(df_from_array.describe(), "\n")  # Статистична інформація про числові колонки

# 5. Індексація (iloc, loc)

# Використання iloc (індексація за позиціями)
print("Вибір рядків за допомогою iloc (індекси):")
print(df_from_array.iloc[1:4])  # Отримання рядків з 1 по 3 за індексами (позиціями), 4 виключено

# Використання loc (індексація за мітками)
print("\nВибір даних за допомогою loc (мітки):")
print(df_from_dict.loc[1:3, ['Film', 'Director']])  # Отримання рядків з мітками 1-3 і тільки колонок 'Film' та 'Director'
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Генерація даних для лінійної функції з шумом
x = np.linspace(0, 10, 100)
true_y = 2 * x + 1
noise = np.random.normal(0, 1, size=x.shape)
y = true_y + noise

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(x, true_y, color='red', label='Теоретична лінія')
plt.scatter(x, y, color='blue', label='Дані з шумом', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Лінійна функція з шумом')
plt.legend()
plt.grid(True)
plt.show()

# Обчислення метрик похибок MAE та MSE
mae = np.mean(np.abs(y - true_y))
mse = np.mean((y - true_y) ** 2)

# Виведення метрик
print(f"MAE: {mae}")
print(f"MSE: {mse}")

# Запис результатів у CSV файл
results = pd.DataFrame({'MAE': [mae], 'MSE': [mse]})
results.to_csv('results.csv', index=False)
```
